```python
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pandas.read_csv('intuit.csv')
```

```python
data.head(10)
```

```python
zip_bins=pandas.get_dummies(data['zip_bins'],drop_first=True,prefix="zip_bins")
```

```python
data_with_dummies=pandas.concat([data,zip_bins],axis=1)
```

We don't want to type the long list of variables over and over again, so I will define this handy list.

```python
varlist=['dollars','zip_bins_2','zip_bins_3','zip_bins_4','zip_bins_5','zip_bins_6',
         'zip_bins_7','zip_bins_8','zip_bins_9','zip_bins_10']
```

Regularized version of the logistic regression takes another parameter, <code>alpha</code>.
This parameter dictacts how strict the model when trowing out variables. High <code>alpha</code> means trowing out
variables more.

```python
res=mba263.logit_reg(data_with_dummies,'res1',varlist,alpha=3)
```

```python
res.summary2()
```

The model threw out <code>zip_bins_10</code> variable. Let's see what happens when we are less strict and set a lower <code>alpha</code>.

```python
res=mba263.logit_reg(data_with_dummies,'res1',varlist,alpha=0.1)
res.summary2()
```

In this case the algorithms kept <code>zip_bins_10</code>. Note: <code>alpha=0</code> is a regular logic (no throwing away variables).
